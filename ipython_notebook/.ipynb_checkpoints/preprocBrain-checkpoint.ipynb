```python
%pylab inline
%config InlineBackend.figure_format = 'svg' 
from sets import Set
from pylab import *
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import sys
import cPickle as pickle


peak_cube = pickle.load(open('preprocBrain.dat', 'rb' ))['peak_cube']
cluster = pickle.load(open('preprocBrain.dat', 'rb' ))['cluster']

Nx, Ny, Nz = peak_cube.shape
```

```python
nCluster = 8

fig = plt.figure(1, (14., 14.))
grid = ImageGrid(fig, 111, # similar to subplot(111)
                nrows_ncols = (2, 4), # creates 2x2 grid of axes
                axes_pad=0.1, # pad between axes in inch.
                )

for i in range(nCluster):
    grid[i].imshow(cluster.reshape(Nx,Ny)==i,cmap='gray') # The AxesGrid object work as a list of axes.
show()

fig.savefig('../figures/preproc/kmeansBrain.eps')
fig.savefig('../figures/preproc/kmeansBrain.jpg')
```

```python
# plot mask
idxBad = [0,3,4,7]
badMask = np.zeros((Nx,Ny,nCluster))
for i in range(nCluster):
    badMask[:,:,i] = cluster.reshape(Nx,Ny)==i

badMask = np.sum(badMask[:,:,idxBad],2)
badMask = badMask==0 #if an element is True, then it is a good pixel
fig = plt.figure()
imshow(badMask)
show

# saving figures
fig.savefig('../figures/preproc/maskBrain.eps')
fig.savefig('../figures/preproc/maskBrain.jpg')

# saving masking data
mask = badMask
with open('mask.dat', 'w') as outfile:
    pickle.dump(mask, outfile, True)
```
